**Outliers**:- An outlier is a data point that differs significantly from other observations.Outliers are detect by using IQR method stands for interquartile range, which is the difference between q3 (75th percentile) and q1 (25th percentile). The IQR method computes lower bound and upper bound to identify outliers.
IQR = Q3 - Q1

```python
Q1=passenger.quantile(0.25)#first quartile
Q3=passenger.quantile(0.75)#third quartile
IQR= Q3- Q1
IQR
```


**Formula for Removing Outliers** = The outliers are the points lying beyond the upper boundary of Q3 +1.5 IQR and the lower boundary of Q1 - 1.5 IQR.

```python
passenger=passenger[~((passenger<(Q1-1.5 * IQR))|(passenger>(Q3+1.5 * IQR))).any(axis=1)]
passenger.shape
# The total 21566 rows are remaining by removing outliers
```

**Model Testing and Training**

Train and Test datasets are the two key concepts of machine learning, where the training dataset is used to fit the model, and the test dataset is used to evaluate the model.

```python
from sklearn.model_selection import train_test_split
```

```python
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=0.2,random_state=1)
# Training dataset 80% and 20% for testing of airline passenger satisification model
```

**Perform Logistic Regression**

**Logistic regression** :- Logistic regression is a data analysis technique that uses mathematics to find the relationships between two data factors. It then uses this relationship to predict the value of one of those factors based on the other. The prediction usually has a finite number of outcomes, like yes or no.

```python
#Perform Logistic Regression
from sklearn.linear_model import LogisticRegression
```

```python
model=LogisticRegression()
model.fit(xtrain,ytrain)
# Fit :-Model fitting is a measure of how well a machine learning model generalizes to similar data to that on which it was trained. A model that is well-fitted produces more accurate outcomes.
```

```python
ypred=model.predict(xtest)
ypred
#Predict :- Prediction” refers to the output of an algorithm after it has been trained on a historical dataset and applied to new data when forecasting the likelihood of a particular outcome
```

**Classification of Model**

Predictive Modeling

The Classification algorithm is a Supervised Learning technique that is used to identify the category of new observations on the basis of training data.his technique can be performed on structured or unstructured data and its main goal is to identify the category or class to which a new data will fall under.

```python
from sklearn.metrics import classification_report, confusion_matrix
```

Dispaly Classification Report

```python
print(classification_report(ytest,ypred))
#A classification report is a performance evaluation metric in machine learning. It is used to show the precision, recall, F1 Score, and support of your trained classification model.
#Precision :- Precision is defined as the ratio of true positives to the sum of true and false positives.
#Recall :- Recall is defined as the ratio of true positives to the sum of true positives and false negatives.
# F1 Score :- The F1 is the weighted harmonic mean of precision and recall. The closer the value of the F1 score is to 1.0, the better the expected performance of the model is.
# Support :- Support is the number of actual occurrences of the class in the dataset. It doesn’t vary between models, it just diagnoses the performance evaluation process
#Note
# If F1 score are high(better) means model are best
```

Display Confusion matrix

```python
print(confusion_matrix(ytest,ypred))
#The confusion matrix is a matrix used to determine the performance of the classification models for a given set of test data.
#The matrix is divided into two dimensions, that are predicted values and actual values along with the total number of predictions.
#Predicted values are those values, which are predicted by the model, and actual values are the true values for the given observations.
#The above  has the following cases:

#True Negative: Model has given prediction No, and the real or actual value was also No.
#True Positive: The model has predicted yes, and the actual value was also true.
#False Negative: The model has predicted no, but the actual value was Yes, it is also called as Type-II error.
#False Positive: The model has predicted Yes, but the actual value was No. It is also called a Type-I error.
```

Display the Training Accuracy of Model

```python
model.score(xtrain,ytrain)
# training accuracy
```

Display the Testing Accuracy of Model

```python
model.score(xtest,ytest)
#testing accuracy
```
